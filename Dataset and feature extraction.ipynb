### add relationships

```python
es = es.add_relationship("users", "user_id", "logons", "user")
es = es.add_relationship("users", "user_id", "devices", "user")
es = es.add_relationship("users", "user_id", "https", "user")
es = es.add_relationship("users", "user_id", "emails", "user")
es = es.add_relationship("users", "user_id", "files", "user")
es
```

### run deep feature synthesis

```python
trans_primitives=['day', 'year', 'month', 'weekday', 'num_words', 'num_characters', 'hour']
trans_primitives
```

```python
agg_primitives=list(ft.primitives.utils.get_aggregation_primitives().keys())
agg_primitives
```

```python
feature_matrix, feature_defs = ft.dfs(entityset=es, target_dataframe_name="users", trans_primitives=trans_primitives, agg_primitives=agg_primitives, max_depth=2)
feature_matrix
```

### remove features with none values

```python
feature_matrix_nona = feature_matrix.dropna(axis=1)
feature_matrix_nona
```

### remove features with string values

```python
str_cols = []
for i, val in enumerate(feature_matrix_nona.iloc[0, :]):
    if isinstance(val, str):
#         print(val)
        str_cols.append(feature_matrix_nona.columns[i])
print(str_cols)
```

```python
feature_matrix_nona_nostr = feature_matrix_nona.drop(columns=str_cols)
feature_matrix_nona_nostr
```

### add labels

```python
insiders_df = pd.read_csv('./dataset/answers/insiders.csv')
insiders_df
```

```python
insiders_df['dataset'].value_counts()
```

```python
insiders_df[insiders_df['dataset'] == 4.2]
```

```python
user_label = {}
for i, row in insiders_df[insiders_df['dataset'] == 4.2].iterrows():
    if row['user'] in user_label:
        print('error in this approach')
    user_label[row['user']] = row['scenario']
```

```python
len(user_label), user_label
```

```python
label = []
for i, row in feature_matrix.iterrows():
    if i in user_label:
        label.append(user_label[i])
    else:
        label.append(0)
```

```python
feature_matrix_nona_nostr['label'] = label
feature_matrix_nona_nostr
```

### save feature matrix

```python
feature_matrix_nona_nostr.reset_index().to_csv("dataset/feature_matrix_nona_nostr.tdt", sep='\t', index=False)
```
